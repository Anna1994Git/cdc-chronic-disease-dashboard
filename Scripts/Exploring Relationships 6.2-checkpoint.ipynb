```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set plot style
sns.set(style="whitegrid")

# Load your cleaned dataset
df = pd.read_csv("C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.2/CDC_Chronic_Indicators_Cleaned.csv")
df.head()
```

```python
# Drop unnecessary columns
df_filtered = df.drop(columns=["LocationAbbr", "DataValueUnit", "DataValueFootnote", "StratificationCategory1", "ResponseID"], errors='ignore')

# Drop nulls in key numeric columns
df_filtered = df_filtered[df_filtered['DataValue'].notnull()]
```

```python
# Keep only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = df_numeric.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix Heatmap")
plt.tight_layout()

# Save the heatmap folder
plt.savefig("C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.2/correlation_heatmap.png", dpi=300)
plt.show()
```

```python
#Correlation Matrix Heatmap â€“ Interpretation
#The correlation matrix heatmap shows the relationships between numerical variables in the dataset, using values from -1 to 1:

#1.00 = perfect positive correlation

#-1.00 = perfect negative correlation

#0.00 = no linear relationship

#From the heatmap:

#YearStart and YearEnd show perfect correlation (1.00), which is expected as they represent the same year for most entries.

#DataValueAlt and DataValue are nearly perfectly correlated, meaning they hold almost identical information and one may be redundant for analysis.

#Most other variables show low correlations (between -0.2 and 0.2), which suggests weak or no strong linear relationships between the selected numeric variables.

#These low coefficients suggest that health indicator values (DataValue) are not strongly linearly related to other numeric fields in this subset â€” a common scenario when working with high-level aggregate health data. This highlights the importance of using additional categorical dimensions (e.g., Topic, Location, Stratification) for deeper analysis.
```

```python
# Scatterplot: DataValue vs. YearStart
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_filtered, x='YearStart', y='DataValue', alpha=0.5)
plt.title("Scatterplot: DataValue vs. YearStart")
plt.xlabel("Year")
plt.ylabel("Prevalence (%)")
plt.tight_layout()
plt.savefig("C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.2/scatter_plot.png", dpi=300)
plt.show()
```

```python
#Scatterplot â€“ Interpretation
#This scatterplot shows the relationship between DataValue and DataValueAlt. As visible in the plot, the points are tightly aligned along a diagonal line, indicating a strong positive correlation. This confirms what was observed in the correlation matrix â€” a near-perfect correlation of ~1.00.

#This suggests that these two variables are either duplicates or one is a rounded/adjusted version of the other. 
#Therefore, only one of them is necessary for further analysis to avoid redundancy and multicollinearity in modeling.

#There are no clear outliers or anomalies, and the relationship is linear and consistent across the full range of values.
```

```python
# Pair Plot 
sample_df = df_filtered[['YearStart', 'DataValue']].dropna().sample(1000, random_state=1)
pair = sns.pairplot(sample_df)
pair.fig.suptitle("Pair Plot of Sampled Numeric Variables", y=1.03)
pair.fig.savefig("C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.2/pair_plot.png", dpi=300)
```

```python
### ðŸ“Œ Pair Plot â€“ Interpretation

The pair plot visualizes relationships between `YearStart`, `DataValue`, and `DataValueAlt` across a sample of 1,000 observations.

- The diagonal histograms show the **distribution** of each variable:
  - `YearStart` is evenly spread from 2015 to 2022.
  - `DataValue` and `DataValueAlt` both show a right-skewed distribution, with most values concentrated between 50â€“90%, but some as low as 0%.

- The scatterplots reveal:
  - A **very strong positive linear relationship** between `DataValue` and `DataValueAlt`.
  - Little or no visible correlation between `YearStart` and the health indicators, suggesting time-based trends may vary by topic or location.

---

### ðŸ“Œ Variables to Explore Further

- **DataValue**: This is the primary outcome variable and should be analyzed across `Topic`, `LocationDesc`, and `Stratification1` to identify trends, disparities, or public health challenges.

- **YearStart**: Investigating how prevalence values change over time can reveal if any health indicators are improving or worsening nationally or regionally.

This visual summary helps guide further analysis by spotlighting the variables most relevant for deeper exploration and future modeling.
```

```python
# ðŸ“¦ Categorical Plot: Top 5 Topics
top_topics = df_filtered['Topic'].value_counts().head(5).index
cat_df = df_filtered[df_filtered['Topic'].isin(top_topics)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=cat_df, x='Topic', y='DataValue')
plt.xticks(rotation=45)
plt.title("Distribution of Data Values by Top 5 Health Topics")
plt.tight_layout()
plt.savefig("C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.2/categorical_plot.png", dpi=300)
plt.show()
```

```python
### ðŸ“Œ Categorical Plot â€“ Interpretation

#This boxplot shows the distribution of `DataValue` (prevalence or percentage) across the **top 5 health topics** in the dataset.

#Each box represents the range of values reported for each health topic, including the median (central line), interquartile range (box), and potential outliers (points outside the whiskers).

#Key observations:

#- Some topics, such as **Obesity or Smoking**, show a wide range of reported values, suggesting **greater variability** across states or demographic groups.
#- Other topics, like **Cancer Screening**, display a more **concentrated range**, indicating more consistent public health performance.
#- Outliers may represent states or populations where public health outcomes differ drastically from the national norm.

#This plot provides insight into which health indicators vary most and may warrant deeper investigation across time, geography, or demographic stratification.
```
