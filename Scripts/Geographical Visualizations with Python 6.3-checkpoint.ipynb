```python
import pandas as pd
import folium
import json
from IPython.display import display


# Load the cleaned data
csv_path = r"C:\Users\annma\OneDrive\CareerFoundary\Achivement 6.3\CDC_Chronic_Indicators_Cleaned_Final.csv"
df = pd.read_csv(csv_path)

# Filter the dataset for Obesity and most recent year
df_obesity = df[df['Topic'] == 'Nutrition, Physical Activity, and Weight Status']
df_obesity = df_obesity[df_obesity['Question'].str.contains('Obesity')]
df_obesity_latest = df_obesity[df_obesity['YearStart'] == df_obesity['YearStart'].max()]

# Group by state and calculate the average DataValue
obesity_avg = df_obesity_latest.groupby('LocationDesc')['DataValue'].mean().reset_index()
obesity_avg.columns = ['STATE', 'VALUE']

# Load GeoJSON file for US states
geo_path = r"C:\Users\annma\OneDrive\CareerFoundary\Achivement 6.3\us-states.json"
with open(geo_path) as f:
    us_states = json.load(f)

# Create a folium map
obesity_map = folium.Map(location=[37.8, -96], zoom_start=4)

# Add the choropleth layer
folium.Choropleth(
    geo_data=us_states,
    name='choropleth',
    data=obesity_avg,
    columns=['STATE', 'VALUE'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Average Obesity Rate (%)',
    nan_fill_color='gray'
).add_to(obesity_map)

# Add layer control
folium.LayerControl().add_to(obesity_map)

# Display the map in Jupyter
obesity_map
obesity_map.save("obesity_choropleth_map.html")
display(obesity_map)

```

```python
import pandas as pd
import folium
import json

# 1. Load the cleaned data
data_path = r"C:\Users\annma\OneDrive\CareerFoundary\Achivement 6.3\CDC_Chronic_Indicators_Cleaned_Final.csv"
df = pd.read_csv(data_path)

# 2. Filter data for diabetes (latest available year)
diabetes_df = df[(df['Topic'] == 'Diabetes') & (df['StratificationCategory1'] == 'Overall')]

# 3. Aggregate average diabetes values by state
diabetes_avg = diabetes_df.groupby("LocationDesc")["DataValue"].mean().reset_index()
diabetes_avg.columns = ["STATE", "VALUE"]

# 4. Load US States GeoJSON file
geo_path = r"C:\Users\annma\OneDrive\CareerFoundary\Achivement 6.3\us-states.json"
with open(geo_path, "r") as file:
    us_states = json.load(file)

# 5. Create the folium map centered on the US
diabetes_map = folium.Map(location=[37.8, -96], zoom_start=4, tiles="cartodbpositron")

# 6. Create the choropleth layer with blue-red color scheme
folium.Choropleth(
    geo_data=us_states,
    name="choropleth",
    data=diabetes_avg,
    columns=["STATE", "VALUE"],
    key_on="feature.properties.name",
    fill_color="RdBu",  # Blue to Red
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Average Diabetes Prevalence (%)",
    nan_fill_color="gray"
).add_to(diabetes_map)

# 7. Save and show the map
map_file_path = r"C:\Users\annma\OneDrive\CareerFoundary\Achivement 6.3\diabetes_choropleth_map.html"
diabetes_map.save(map_file_path)

# Display in notebook
diabetes_map
```
