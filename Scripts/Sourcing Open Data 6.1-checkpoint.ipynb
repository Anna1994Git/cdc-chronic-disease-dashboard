```python
import pandas as pd
import numpy as np
import os
```

```python
# Local path to dataset
file_path = "C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.1/U.S._Chronic_Disease_Indicators.csv"

# Load the CSV file
df = pd.read_csv(file_path)
print("Original shape:", df.shape)
df.head()
```

```python
# Drop fully empty columns
empty_cols = [col for col in df.columns if df[col].isnull().all()]
df_cleaned = df.drop(columns=empty_cols)
```

```python
print("Empty columns dropped:", empty_cols)
```

```python
# Show shape before removing duplicates
print("Shape before removing duplicates:", df_cleaned.shape)

# Drop duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Show shape after removing duplicates
print("Shape after removing duplicates:", df_cleaned.shape)

#how many rows were removed
duplicates_removed = df.shape[0] - df_cleaned.shape[0]
print("Number of duplicate rows removed:", duplicates_removed)
```

```python
# Show data types before conversion
print("Data types before conversion:\n")
print(df_cleaned.dtypes)

# Convert year columns to integers
df_cleaned['YearStart'] = pd.to_numeric(df_cleaned['YearStart'], errors='coerce').astype("Int64")
df_cleaned['YearEnd'] = pd.to_numeric(df_cleaned['YearEnd'], errors='coerce').astype("Int64")

# Convert data value to numeric (some may be stored as strings)
df_cleaned['DataValue'] = pd.to_numeric(df_cleaned['DataValue'], errors='coerce')

# Optionally: Convert geographic codes to string
df_cleaned['LocationAbbr'] = df_cleaned['LocationAbbr'].astype(str)

# Show data types after conversion
print("\nData types after conversion:\n")
print(df_cleaned.dtypes)

# Check how many values were successfully converted (no longer NaN)
print("\nNon-null values after conversion:")
print(df_cleaned[['YearStart', 'YearEnd', 'DataValue']].notnull().sum())
```

```python
## Step 3: Summary & Data Profile
```

```python
# Show data types
print("\nData types:")
print(df_cleaned.dtypes)

# Show missing values per column
print("\nMissing values:")
print(df_cleaned.isnull().sum().sort_values(ascending=False))

# Summary stats for numerical data
print("\nSummary statistics:")
print(df_cleaned.describe())

# Unique counts for key categorical variables
print("\nUnique values in selected columns:")
print(df_cleaned[['LocationDesc', 'YearStart', 'Topic', 'Stratification1']].nunique())
```

```python
## Step 4: Save Cleaned Dataset
output_path = "C:/Users/annma/OneDrive/CareerFoundary/Achivement 6.1/CDC_Chronic_Indicators_Cleaned.csv"
df_cleaned.to_csv(output_path, index=False)
print(f"\nCleaned dataset saved to:\n{output_path}")
```
